## Data access
---
You can call the data from `get_data` after importing using the following convention:
> `get_data.what` where "what" is the data you want such as X_train or y_test
- **Available sets**
    - **X_train** - ecg data, for corresponding ecg_id see y_train at the same *numeric index*. For full meta-data, get ecg_id from y_train then query `Y` at that ecg_id. 
    - **X_test** - same as above, test split.
    - **y_train** - ecg_id and true/false label of normal (true if normal ecg, false otherwise)
    - **y_test** - same as above, test split.
    - **X** - full ecg data for fold 9 of master data.
    - **Y** - full meta-data for fold 9 of master data.
